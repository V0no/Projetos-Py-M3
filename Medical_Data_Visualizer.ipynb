```python
# --- setup para exibir gr√°ficos no Colab ---
%matplotlib inline
import os, pandas as pd, numpy as np, seaborn as sns, matplotlib.pyplot as plt
from IPython.display import display

# 1) Verificar se o CSV est√° no diret√≥rio atual
print("Arquivos no diret√≥rio atual:", os.listdir())
assert os.path.exists("medical_examination.csv"), "Arquivo medical_examination.csv n√£o encontrado no diret√≥rio atual."

# 2) Carregar dados
df = pd.read_csv("medical_examination.csv")
print("Shape do df bruto:", df.shape)

# 3) Feature engineering conforme enunciado
bmi = df["weight"] / ((df["height"] / 100) ** 2)
df["overweight"] = (bmi > 25).astype(int)
df["cholesterol"] = (df["cholesterol"] > 1).astype(int)
df["gluc"]        = (df["gluc"] > 1).astype(int)

def draw_cat_plot():
    # Melt ‚Üí groupby ‚Üí contagem
    df_cat = pd.melt(
        df,
        id_vars=["cardio"],
        value_vars=["cholesterol", "gluc", "smoke", "alco", "active", "overweight"]
    )
    df_cat = (
        df_cat.groupby(["cardio", "variable", "value"], as_index=False)
              .size()
              .rename(columns={"size": "total"})
    )
    g = sns.catplot(
        data=df_cat,
        x="variable",
        y="total",
        hue="value",
        col="cardio",
        kind="bar"
    )
    fig = g.fig
    fig.savefig("catplot.png", bbox_inches="tight")
    return fig

def draw_heat_map():
    # Limpeza
    df_heat = df[
        (df["ap_lo"] <= df["ap_hi"]) &
        (df["height"] >= df["height"].quantile(0.025)) &
        (df["height"] <= df["height"].quantile(0.975)) &
        (df["weight"] >= df["weight"].quantile(0.025)) &
        (df["weight"] <= df["weight"].quantile(0.975))
    ].copy()
    print("Shape ap√≥s limpeza (heatmap):", df_heat.shape)

    corr = df_heat.corr(numeric_only=True)
    mask = np.triu(np.ones_like(corr, dtype=bool))

    fig, ax = plt.subplots(figsize=(12, 10))
    sns.heatmap(
        corr, mask=mask, annot=True, fmt=".1f",
        center=0, square=True, linewidths=.5,
        cbar_kws={"shrink": .5}, ax=ax
    )
    fig.savefig("heatmap.png", bbox_inches="tight")
    return fig

# --- gerar e MOSTRAR as figuras explicitamente ---
plt.close('all')                 # evita gr√°ficos ‚Äúfantasma‚Äù de c√©lulas anteriores
fig1 = draw_cat_plot()
display(fig1)                    # garante exibi√ß√£o no output da c√©lula

fig2 = draw_heat_map()
display(fig2)                    # garante exibi√ß√£o no output da c√©lula

print("Arquivos salvos:", [f for f in os.listdir() if f.endswith('.png')])
```
